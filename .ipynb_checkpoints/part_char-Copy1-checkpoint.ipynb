```python
import pandas as pd
import numpy as np
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import lxml
import json
import time
import csv
import random
```

```python
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
```

```python
service = Service(executable_path=ChromeDriverManager().install())
```

```python
# Загружаем датафрейм ссылок

lnk_csv = 'https://raw.githubusercontent.com/ArtemDorofeev/Kinopoisk/main/csv/df_2015-2022.csv'
data = pd.read_csv(lnk_csv)
df_full = pd.DataFrame(data)
```

```python
df = df_full[['name', 'links', 'year', 'type']]
df
```

```python
# Формируем суп-объект

def get_soup(url):    
    driver.get(url)    
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")
    #info_section = soup.select('div.styles_root__2kxYy')
    return soup
```

```python
# Собираем характеристики фильмов

def get_top(row):
    list_top = []
    try:
        name = row.select_one('h1[class*="styles_title"]').get_text()
    except:
        name = 'none'
    
    try:
        description = row.select_one('div[class*="styles_topText"]').get_text()
    except:
        description = 'none'
        
    try:
        rating = row.select_one('span[class*="film-rating-value"]').get_text()
    except:
        rating = 'none'
        
    try:
        grades = row.select_one('div[class*="styles_countBlock"]').get_text()
    except:
        grades = 'none'
    
    dic = {"name": name, "description": description, "rating": rating, "grades": grades}
    list_top.append(dic)
    return list_top

def get_about(row):    
    about_block = row.select_one('div[data-test-id*="encyclopedic-table"]')
    about_row = about_block.select('div[class*="styles_row"]')
    list_about = []
    for i in about_block:
        title = i.select_one('div[class*="styles_title"]').get_text()
        val_block = i.select_one('div[class*="styles_value"]')
        if val_block.select('a[class*="styles_link"]'):
            values = []
            val_links = val_block.select('a[class*="styles_link"]')
            for j in val_links:
                link = j.get('href')
                val = j.get_text()
                lst = [val, link]
                values.append(lst)
        else:
            values = val_block.get_text()
        dic = {title: values}
        list_about.append(dic)        
    return list_about


def get_actor(row):
    list_actor = []
    try:        
        actor_block = row.select('div[class*="styles_actors"] ul li')        
        for i in actor_block:
            link = i.select_one('a').get('href')
            actor_name = i.get_text()
            dic = {"actor_name": actor_name, "link": link}
            list_actor.append(dic)
    except:
        list_actor = list_actor
    return list_actor

def get_low(row):    
    try:
        critic_section = row.select_one('div[class*="styles_criticRatingSection"]')    
        try:
            critic_positive = critic_section.select_one('span[class*="film-rating-value"]').get_text()
        except:
            critic_positive = 'none'
        try:   
            critic_value = critic_section.select_one('div[class*="styles_countBlock"]').get_text()
        except:
            critic_value = 'none'
        try:
            critic_star = critic_section.select_one('div[class*="styles_starValue"]').get_text()
        except:
            critic_star = 'none'
        lst = [critic_positive, critic_value, critic_star]
    except:
        lst = []
    return lst
    
```

```python
# Собираем характеристики в общий словарь

def full_parse(row):    
    #time.sleep(tm)
    try:
        soup = get_soup(row)
        if soup.select_one('div[class*="CheckboxCaptcha"]'):
            time.sleep(20)
        else:
            name_rate = get_top(soup)
            about_film = get_about(soup)
            actors = get_actor(soup)
            critic_rate = get_low(soup)
            lst = [name_rate, about_film, actors, critic_rate]
    except:
        lst = []
    return lst
```

```python
driver = webdriver.Chrome(service=service) # Открывает новую отдельную вкладку браузера хром от selenium
```

```python
# ЗАПУСК ПАРСЕРА!!!

n = 1000 # размер партиции
m = 10 # Количество партиций
path = 'D:\\Data\\Kinopoisk\\characteristics\\kino_char_part{}.csv'
x = 11000
for j in range(1, m+1):    
    dsa = df.loc[x:x+n-1]
    x += n    
    dsa['char'] = dsa['links'].apply(full_parse)
    dsa.to_csv(path.format(x))
```
