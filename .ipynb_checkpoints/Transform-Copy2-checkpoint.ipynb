```python
import pandas as pd
import numpy as np
import json
import csv
import os
```

```python
dirs = os.listdir(path="D:\\Data\\Kinopoisk\\characteristics\\new") 
```

```python
lnk = 'D:\\Data\\Kinopoisk\\characteristics\\new\\{}'
dataset = pd.DataFrame()
for i in dirs:
    lst = lnk.format(i)
    fls = os.listdir(path=lst)
    for j in fls:
        d = pd.read_csv(lst+'\\'+j)
        dataset = pd.concat([dataset, d], ignore_index=True)
```

```python
df = dataset.copy()
df.sample(10)
```

```python
df = df[df['char'] != '[]']
```

```python
df['charlist'] = df['char'].apply(lambda x: eval(x))
```

```python
df.sample(10)
```

```python
df.to_csv('D:\\Data\\Kinopoisk\\characteristics\\char100_250.csv')
```

```python
df['full_name'] = df['charlist'].apply(lambda x: x[0][0]['name'])
df['rating'] = df['charlist'].apply(lambda x: x[0][0]['rating'])
df['quant_grad'] = df['charlist'].apply(lambda x: x[0][0]['grades'])
df['description'] = df['charlist'].apply(lambda x: x[0][0])
```

```python
df['charlist'][44460][1]
```

```python
def keys_unique(row):
    lst = []
    dcts = row[1]
    for i in dcts:
        k = list(i)
        lst += k
    return lst
```

```python
keys_unique(df['charlist'][44460])
```

```python
df['key_char'] = df['charlist'].apply(keys_unique)
```

```python
x = df['key_char'].sum()
```
