```python
import pandas as pd
import numpy as np
import json
import csv
import os
```

```python
dirs = os.listdir(path="D:\\Data\\Kinopoisk\\CSV\\characteristics")
```

```python
lnk = 'D:\\Data\\Kinopoisk\\CSV\\characteristics\\{}'
df = pd.DataFrame()
for i in dirs:
    lst = lnk.format(i)
    fls = os.listdir(path=lst)
    for j in fls:
        d = pd.read_csv(lst+'\\'+j)
        df = pd.concat([df, d], ignore_index = True)
    
```
